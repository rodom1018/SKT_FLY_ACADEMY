```python
seoul_map = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', zoom_start = 12)
seoul_map
```

```python
df = pd.read_csv('./samples/seoul_hospital.csv', encoding = 'cp949')
# https://www.data.go.kr/data/15088910/fileData.do
# 서울특별시 응급실 위치 정보 . 
for name, lat, lng in zip(df.index, df.병원위도, df.병원경도):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)
    
seoul_map
```

```python
seoul_map2 = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', zoom_start = 12)
seoul_map2
```

```python
df = pd.read_csv('./samples/seoul_hospital.csv', encoding = 'cp949')
# https://www.data.go.kr/data/15088910/fileData.do
# 서울특별시 응급실 위치 정보 . 
for name, lat, lng in zip(df.index, df.병원위도, df.병원경도):
    folium.CircleMarker([lat, lng],
                      radius = 10,
                      color = 'brown',
                      fill = True,
                      fill_color = 'coral',
                      fill_opacity = 0.7,
                      popup = name).add_to(seoul_map2)
    
seoul_map2
```
