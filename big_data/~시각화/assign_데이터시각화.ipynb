```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os
import pandas as pd
```

```python
csv = os.path.join('./samples/crime.csv')
df = pd.read_csv(csv, encoding = 'cp949')
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(6,6))

ax= fig.add_subplot(111)
ax.plot(df.index, df['2014'], label='2014')
#ax.plot(df.index, df['2015'], label='2015')
ax.plot(df.index, df['2016'], label='2016')
#ax.plot(df.index, df['2017'], label='2017')
ax.plot(df.index, df['2018'], label='2018')

ax.set_title('Monthly Statistics on Illegal Parking Blocks in Sasang-gu, Busan')
ax.set_ylabel('number of people')
ax.set_xlabel('month')
ax.legend()
plt.show()
```
