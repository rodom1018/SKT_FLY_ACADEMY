```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='NanumGothic')

csv = os.path.join('./samples/4816.csv')
df = pd.read_csv(csv, index_col=0 , parse_dates=['Date'])

fig = plt.figure(figsize = (10,4))
ax1 = fig.add_subplot(111)

ax1.plot(df.index, df['Close'], color='b', label='주가')

ax2 = ax1.twinx()
ax2.bar(df.index, df['Volume'], color='g', label='거래총액', width=2)

ax1.set_xlabel('년월')
#ax1.set_yticks([i * 500 for i in range(5)])
ax1.set_ylabel('주가')
ax2.set_ylabel('거래총액')
ax1.set_title('주가와 거래총액')
ax1.legend(loc=1)
ax2.legend(loc=2)
```

```python
import os
import pandas as pd
csv = os.path.join("./samples/anime_master.csv")
df = pd.read_csv(csv, index_col = 'anime_id')
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(df['members'], df['rating'], alpha = 0.5)
plt.show()
```

```python
df.loc[df['members']>=800000, ['name', 'members']]
```

```python
df.loc[(df['members'] >= 600000) & (df['rating']>=8.5), ['name', 'members']]
```

```python
types = df['type'].unique()
types
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

for t in types:
    x = df.loc[df['type'] == t, 'members']
    y = df.loc[df['type'] == t, 'rating']
    ax.scatter(x, y, alpha=0.5, label=t)
    
ax.set_title('배급 종별로 그룹화한 데이터 산포도 그래프')
ax.set_xlabel('Members')
ax.set_ylabel('Rating')
ax.legend(loc='lower right', fontsize=12)

plt.show()
```

```python
csv = os.path.join('samples/anime_genre_top10_pivoted.csv')
dfag = pd.read_csv(csv, index_col = 'genre')
dfag
```

```python
plt.rc('font', family = 'NanumGothic')
plt.rc('axes', unicode_minus = False)
```

```python
fig = plt.figure(figsize = (18,3))
ax = fig.add_subplot(111)

wt = np.array(range(len(dfag)))
w = 0.1

for i in dfag.columns:
    ax.bar(wt, dfag[i],width=w , label = i)
    wt = wt + w

ax.set_xticks(np.array(range(len(dfag))))
ax.set_xticklabels(dfag.index, ha='left')
ax.set_ylabel('누적 멤버수')
ax.set_yscale('log') # 한개가 너무 압도적이면 scale을 줄여서 ... 나머지가
                    #잘보이게 함
ax.legend()

plt.show()
```

```python
import os
import pandas as pd
csv = os.path.join("./samples/anime_master.csv")
df = pd.read_csv(csv, index_col="anime_id")
df.head()
```
